# <p style="background-color:#dcf0f2;font-family:newtimeroman;color:#2e5777;font-size:150%;text-align:center;border-radius:40px 40px;">Data Preprocessing</p>

```python
# List of columns to be label encoded
columns = ['Gender',
            'family_history_with_overweight',
            'FAVC',
            'CAEC',
            'SCC',
            'NObeyesdad'] 

# Initialize the LabelEncoder object
L_E = LabelEncoder()    

# For each specified column, apply label encoding to transform categorical data to numerical
for col in columns:
    df[col] = L_E.fit_transform(df[[col]].values.ravel())
```

```python
# Extract values of the 'Weight' column from the DataFrame
weight = df['Weight'].values

# Extract values of the 'Gender' column from the DataFrame
gender = df['Gender'].values

# Extract values of the 'family_history_with_overweight' column from the DataFrame
obesity_history = df['family_history_with_overweight'].values

# Extract values of the 'FAF' column from the DataFrame
faf = df['FAF'].values

# Extract values of the 'FAVC' column from the DataFrame
favc = df['FAVC'].values

# Extract values of the 'CAEC' column from the DataFrame
caec = df['CAEC'].values

# Extract values of the 'SCC' column from the DataFrame
scc = df['SCC'].values

# Create an array X with the selected feature columns, transposed
X = np.array([weight, gender, obesity_history, faf, favc, caec, scc]).T

# Extract and ravel the target column values
y = df['NObeyesdad'].values.ravel()

# Split the dataset into training and testing sets with a test size of 20% and a fixed random state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13684)
```

# <p style="background-color:#dcf0f2;font-family:newtimeroman;color:#2e5777;font-size:100%;text-align:center;border-radius:40px 40px;">Metrics</p>

```python
def GeneralMetrics(y_true, y_pred, y_prob):
    print(f'Precision (macro): {metrics.precision_score(y_true, y_pred, average="macro")}\n')
    print(f'Accuracy: {metrics.accuracy_score(y_true, y_pred)}\n')
    print(f'Recall (macro): {metrics.recall_score(y_true, y_pred, average="macro")}\n')
    print(f'F1 Score (macro): {metrics.f1_score(y_true, y_pred, average="macro")}\n')
    print(f'Confusion Matrix:\n {metrics.confusion_matrix(y_true, y_pred)}\n')
    print(f'AUC (macro): {metrics.roc_auc_score(y_true, y_prob, average="macro", multi_class="ovr")}\n')
```

# <p style="background-color:#dcf0f2;font-family:newtimeroman;color:#2e5777;font-size:150%;text-align:center;border-radius:40px 40px;">Model Selection</p>

```python
def BasicAlgorithms(X_train, y_train):
    models = {
    'LogisticRegression': LogisticRegression(n_jobs=-1),
    'SVC': SVC(random_state=8452),
    'DecisionTreeClassifier': DecisionTreeClassifier(random_state=8452),
    'RandomForestClassifier': RandomForestClassifier(random_state=8452, n_jobs=-1),
    'GradientBoostingClassifier': GradientBoostingClassifier(random_state=8452),
    'AdaBoostClassifier': AdaBoostClassifier(random_state=8452),
    'KNeighborsClassifier': KNeighborsClassifier(n_jobs=-1),
    'GaussianNB': GaussianNB(),}
    model_scores = []
    for name, model in models.items():
    # Perform cross-validation
        cv_scores = cross_val_score(model, X_train, y_train, cv=7, scoring='accuracy')
    
    # Calculate evaluation metrics
        accuracy = cv_scores.mean()

    # Append the metrics to model_scores
        model_scores.append((name, accuracy))

    # Create a DataFrame from model_scores
    columns = ['Model', 'Accuracy',]
    df_model_scores = pd.DataFrame(model_scores, columns=columns)

    # Sort the DataFrame by Accuracy in descending order
    df_model_scores = df_model_scores.sort_values(by='Accuracy', ascending=False)

    # Display the DataFrame
    print(df_model_scores)
BasicAlgorithms(X_train, y_train) 
```

### Conclusion:
After careful consideration, I will use Random Forest as our final algorithm for model training. While other algorithms may have shown slightly higher accuracy, Random Forest's interpretability, lower risk of overfitting, parallelization capabilities, robustness, and reduced hyperparameter tuning requirements make it the most suitable choice for our specific use case.

```python
# Define the RandomForestModel function
def RandomForestModel(X_train, X_test, y_train, y_test):
    # Initialize the RandomForestClassifier with a fixed random state
    rf = RandomForestClassifier(random_state=8452)

    # Define the parameter grid for GridSearchCV
    param_grid = {
        'n_estimators': [125, 250, 500],
        'max_depth': [5, 7, 9],
        'min_samples_split': [5, 7, 9]}
    
    # Initialize GridSearchCV with the parameter grid, 7-fold cross-validation, and parallel processing
    rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=7, n_jobs=-1)

    # Fit the model to the training data
    rf.fit(X_train, y_train)
    
    # Print the best hyperparameters found by GridSearchCV
    print(rf.best_params_)
    
    # Retrieve the best estimator (model) found by GridSearchCV
    rf = rf.best_estimator_
    
    # Predict the labels of the test set
    y_pred = rf.predict(X_test)
    
    # Predict the probabilities of the test set
    y_prob = rf.predict_proba(X_test)

    # Perform cross-validation and calculate AUC scores
    cv_scores = cross_val_score(rf, X, y, cv=7, scoring='roc_auc_ovr')
    # Print the mean AUC score from cross-validation
    print(f'\nAUC cross validation mean: {cv_scores.mean()}\n')

    # Return general metrics of the model's performance
    return GeneralMetrics(y_test, y_pred, y_prob)

# Call the RandomForestModel function with the training and testing datasets
RandomForestModel(X_train, X_test, y_train, y_test)
```

### Model Analysis:
We have a fairly balanced model that can differentiate well between each of the classes. Remember that we have 7 different classes, so the model could easily get confused during training. The model has predicted 83.25% of the cases correctly, however this is a general average among all classes. If we look at the main confusion matrix, we see how even some classes have been classified correctly 100%, and then other classes around 90%, and in 2 other classes is where we see a considerably lower performance, with approximately 70%. Even so, it is a balanced model that has no type of bias when predicting any data. Proof of this is the percentage difference between precision and recall (0.8346 & 0.8291) respectively.
